```python
a=5
b=77
c=6
g=5

print(id(a))
print(id(b))
print(id(c))
print(id(g))

class test:
    def __init__(self,v):
        self.v = v
    def __str__(self):     
       return str(self.v)
a = test(2)      
b = test(2)
print(a)
print(b)
print(id(a))
print(id(b))
```
