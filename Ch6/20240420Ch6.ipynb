```python
def sumValue(*args):
    v=0
    for a in args:
        v+=a
    return v
print(sumValue())
print(sumValue(2))
print(sumValue(2,3))
print(sumValue(2,3,4))
```

```python
def calculate(*values,**args):
    cType="+"
    if "cType" in args:
        cType=args["cType"]
    result=0
    if cType =="*":
        result=1
    for v in values:
        if cType =="+":
            result+=v
        elif cType =="*":
            result*=v
    return result
calculate(3,2,4,cType="*")
```
