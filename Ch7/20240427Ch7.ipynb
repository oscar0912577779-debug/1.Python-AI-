## 繼承

````python
class Animal:
    def __init__(self,name,age=0,atk=2,agi=3,mp=6):
        self.name=name
        self.age=age
        self.atk=atk
        self.agi=agi
        self.mp=mp
    def printInfo(self):
        print("資訊:",self.name,self.age,self.atk,self.agi,self.mp)
an1=Animal("Bobo")
an1.printInfo()
an2=Animal("Lulu",2)
an2.printInfo()
````

````python
class Cat(Animal):
    def printInfo(self):
        print("c:",self.name,self.age)
cat1=Cat("Kitty",3)
cat1.printInfo()
````

````python
class Dog(Animal):
    def printInfo(self):
        print("D:",end="")
        #super()呼叫父類別的東西
        super().printInfo()
        
dog1=Dog("Kiki",2)
dog1.printInfo()
````

````python
class S:
    def method1(self):
        print("S method1")
    def method2(self):
        print("S method2")
class A(S):
    def method3(self):
        print("A method3")
class B(S):
    def method3(self):
        print("B method2")
    def method2(self):
        print("B method3")
class C(A,B):
    def method4(self):
        print("C method4")
c=C()
c.method4()
c.method3()
c.method2()
c.method1()
````

````python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):
        return f"({self.x},{self.y})"
    def __add__(self,other):
        x=self.x+other.x
        y=self.y+other.y
        return Point(x,y)
    def __len__(self):
        myLen=(self.x**2)+(self.y**2)
        return myLen
    def __lt__(self,other):
        selfLen= len(self)
        otherLen=len(other)
        return selfLen<otherLen
p1=Point(2,3)
p2=Point(-1,2)
p3=Point(9,11)
p4=Point(7,12)

print(p1<p2)
print(p1>p2)
pointList=[p1,p2,p3,p4]
pointList.sort()
for p in pointList:
    print(p)
````
