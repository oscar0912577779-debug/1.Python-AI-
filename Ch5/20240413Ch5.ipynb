# 作業 文字接龍

```python
print("五次輸入就會退出遊戲")
txt=input("請輸入一個字串")
strList=[]
strList.append(txt)
msg=txt
count=0
print(len(txt))
while count<5:
    print(f"上一個字串是:{msg}")
    tmpTxt=input(f"請輸入-{txt[-1]}-開始的字串")
    if tmpTxt[0]==txt[-1]:
        txt=tmpTxt
        strList.append(tmpTxt)
        msg="-".join(strList)
    else:
        count+=1
```

# key與value

```python
d2={"a":100,"b":200,"c":300}
for k in d2: #預設取key而非key跟value
    print(f"{k}-{d2[k]}")
print()
for k2 in d2.keys():
    print(f"{k2}-{d2[k2]}")
print()
for v in d2.values():
    print(f"v:{v}")
print()
for v in d2.values():
    print(v)
print()    
for k in d2:
    print(k)
```

```python
d3={"a":100,"b":200,"c":300}
k="d"
findKey=False #預設沒找到key
for key in d3:
    if key==k:
        findKey=True
        break
#找不到key顯示無此key
if findKey:
    print(d3[k])
else:
    print("顯示無此key")
```

```python
myList=["木瓜","西瓜","蘋果","香蕉"]
a=input("水果名稱")
findFruit=False
for i in myList:
    if i ==a:
        findFruit=True
        break
if findFruit:
    print(f"{a}水果 在清單中")
else:
    print(f"{a}水果 不在清單中")
```
