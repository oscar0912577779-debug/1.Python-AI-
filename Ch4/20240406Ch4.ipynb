# pop
## 如果pop()內沒有參數，則移除最後一筆資料。
## pop()內若有參數，則移除指定位置的資料。

```python
list2=[]
list2.append("Ken")
list2.append("Viven")
list2.append("Lucy")
list2.append("Iris")
print(list2)
print(list2.pop())
print(list2.pop())
print(list2.pop())
print(list2.pop())

#print(list2.pop())   #list空了出錯
```

```python
list2=[]
list2.append("Ken")
list2.append("Viven")
list2.append("Lucy")
list2.append("Iris")
while len(list2)>0:
    print(list2.pop(0))
```
