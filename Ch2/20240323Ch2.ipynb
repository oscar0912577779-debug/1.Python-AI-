```python
bmi=100
if  bmi < 0:
    print("錯誤")
elif bmi >= 0 and bmi <= 18.5:
    print("太輕")
elif bmi > 18.5 and bmi <= 25:
    print("正常")
elif bmi > 25 and bmi <= 30:
    print("過重")
else:
    print("肥胖")
#<18.5 太輕
#18.5~25 正常
#25~30 過重
#>30肥胖
#BMI=體重/身高^2(單位公尺)
```

# break代表不回頭，跳出迴圈。
# continue 代表繼續執行下一個迴圈。

```python
for i in range(11):
    print(i)
    if i==5:
        break
```

```python
for i in range(1,11):
    if i==5 or i ==7:
        continue
    print(i,end=" ")
```
