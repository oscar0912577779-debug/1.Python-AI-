# find

```python
str1 = "This is Python.That is Java,This is SQLite"
print(str1)
search1 = "Python"
print(str1.find(search1))
search2 = "not"
print(str1.find(search2))
search3 = "t"
print(str1.find(search3))
search4 = "is"
print(str1.find(search4,7))
```

```python
str1 = "This is Python.That is Java,This is SQLite"
search1 = "is"
index = 0
while True:
    index = str1.find(search1,index)
    if index == -1:
        break
    print("is:",index)
    index+=1
```
